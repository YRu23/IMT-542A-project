```python
import pandas as pd

# Load the CSV files into DataFrames
file_path_main = '/content/movie_score_imdb_tmdb.csv'
file_path_metacritic = '/content/metacritic_movies.csv'

df_main = pd.read_csv(file_path_main)
df_metacritic = pd.read_csv(file_path_metacritic)

# Inspect the columns of each DataFrame
print("Columns in movie_score_imdb_tmdb.csv:")
print(df_main.columns)

print("\nColumns in metacritic_movies.csv:")
print(df_metacritic.columns)

# Display the first few rows of each DataFrame to understand their structure
print("\nFirst few rows of movie_score_imdb_tmdb.csv:")
print(df_main.head())

print("\nFirst few rows of metacritic_movies.csv:")
print(df_metacritic.head())

# Merge the DataFrames on 'movie_title'
final_merged_df = pd.merge(df_main, df_metacritic[['movie_title', 'metascore']], on='movie_title', how='left')

# Fill NaN values in 'metascore' with 0
final_merged_df['metascore'].fillna(0, inplace=True)

# Convert 'metascore' to integer
final_merged_df['metascore'] = final_merged_df['metascore'].astype(int)

# Keep only the required columns: 'imdb id', 'movie_title', 'Year', 'genres', 'director1', 'director2', 'IMDb score', 'tmdb score', 'metascore'
final_df = final_merged_df[['imdb id', 'movie_title', 'Year', 'genres', 'director1', 'director2', 'IMDb score', 'tmdb score', 'metascore']]

# Display the final DataFrame
print("Final DataFrame:")
print(final_df.head())

# Save the final DataFrame to a new CSV file named 'movie_rating_imdb_tmdb_metacritic.csv'
final_df.to_csv('movie_rating_imdb_tmdb_metacritic.csv', index=False)

# Download the final CSV file
files.download('movie_rating_imdb_tmdb_metacritic.csv')
```
