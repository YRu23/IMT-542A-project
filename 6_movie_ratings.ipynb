```python
import pandas as pd

# Load the CSV file into a DataFrame
file_path = '/content/movie_rating_imdb_tmdb_metacritic.csv'
df = pd.read_csv(file_path)

# Inspect the columns of the DataFrame
print("Columns in movie_rating_imdb_tmdb_metacritic.csv:")
print(df.columns)

# Display the first few rows of the DataFrame to understand its structure
print("\nFirst few rows of movie_rating_imdb_tmdb_metacritic.csv:")
print(df.head())

# Normalize the Metacritic scores to a common scale (0-10)
df['metascore_normalized'] = df['metascore'] / 10

# Calculate the average score from IMDb, TMDb, and normalized Metacritic
df['average_score'] = df[['IMDb score', 'tmdb score', 'metascore_normalized']].mean(axis=1)

# Round the average score to one decimal place
df['average_score'] = df['average_score'].round(1)

# Display the DataFrame with the average score
print("\nDataFrame with Average Score:")
print(df[['imdb id', 'movie_title', 'IMDb score', 'tmdb score', 'metascore', 'metascore_normalized', 'average_score']].head())

# Save the DataFrame with the average score to a new CSV file
df.to_csv('movie_ratings.csv', index=False)

# Download the final CSV file
files.download('movie_ratings.csv')
```
