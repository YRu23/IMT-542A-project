```python
import pandas as pd

# Load the CSV file into a DataFrame
file_path = '/content/title.crew.csv'
df_title_crew = pd.read_csv(file_path)

# Inspect the columns of the DataFrame
print("Columns in title.crew.csv:")
print(df_title_crew.columns)

# Display the first few rows of the DataFrame to understand its structure
print("\nFirst few rows of title.crew.csv:")
print(df_title_crew.head())

# Check for rows with multiple directors
print("\nRows with multiple directors:")
multiple_directors = df_title_crew[df_title_crew['directors'].str.contains(',')]
print(multiple_directors)

# Separate the values in the 'directors' column into 'nconst' and 'nconst2'
df_title_crew[['nconst', 'nconst2']] = df_title_crew['directors'].str.split(',', n=1, expand=True)

# Drop the original 'directors' column
df_title_crew = df_title_crew.drop(columns=['directors'])

# Display the cleaned DataFrame
print("\nCleaned DataFrame:")
print(df_title_crew.head())
```

```python
# Save the cleaned DataFrame to a new CSV file
df_title_crew.to_csv('cleaned_title_crew.csv', index=False)

# Download the cleaned CSV file
files.download('cleaned_title_crew.csv')
```

Merge the titl.crew.csv with the nam.basics.csv

```python
import pandas as pd

# Load the CSV files into DataFrames
file_path_crew = '/content/cleaned_title_crew.csv'  # Adjust the path if needed
file_path_basics = '/content/nam.basics.csv'  # Adjust the path if needed

df_title_crew = pd.read_csv(file_path_crew)
df_basics = pd.read_csv(file_path_basics)

# Inspect the columns of each DataFrame
print("Columns in cleaned_title_crew.csv:")
print(df_title_crew.columns)

print("\nColumns in nam.basics.csv:")
print(df_basics.columns)

# Display the first few rows of each DataFrame to understand their structure
print("\nFirst few rows of cleaned_title_crew.csv:")
print(df_title_crew.head())

print("\nFirst few rows of nam.basics.csv:")
print(df_basics.head())

# Rename columns in df_title_crew
df_title_crew = df_title_crew.rename(columns={'nconst2': 'nconst_2'})

# Merge the DataFrames to replace nconst with primaryName for the first director
df_merged1 = pd.merge(df_title_crew, df_basics[['nconst', 'primaryName']], left_on='nconst', right_on='nconst', how='left')
df_merged1 = df_merged1.rename(columns={'primaryName': 'director1'})

# Merge the DataFrames to replace nconst_2 with primaryName for the second director
df_merged_final = pd.merge(df_merged1, df_basics[['nconst', 'primaryName']], left_on='nconst_2', right_on='nconst', how='left', suffixes=('', '_2'))
df_merged_final = df_merged_final.rename(columns={'primaryName': 'director2'})

# Keep only the required columns: 'tconst', 'director1', 'director2'
final_df = df_merged_final[['tconst', 'director1', 'director2']]

# Display the final DataFrame
print("Final DataFrame:")
print(final_df.head())
```

```python
# Save the final DataFrame to a new CSV file
final_df.to_csv('final_directors_v2.csv', index=False)

# Download the final CSV file
files.download('final_directors_v2.csv')
```
