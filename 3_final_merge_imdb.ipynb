```python
from google.colab import files

# Upload the files
uploaded = files.upload()

# List the uploaded files
for filename in uploaded.keys():
    print(f'User uploaded file "{filename}" with length {len(uploaded[filename])} bytes')
```

```python
import pandas as pd

# Load the CSV files into DataFrames
file_path_movies = '/content/final_merged_movies_ratings.csv'
file_path_directors = '/content/final_directors_v2.csv'

df_movies = pd.read_csv(file_path_movies)
df_directors = pd.read_csv(file_path_directors)

# Inspect the columns of each DataFrame
print("Columns in final_merged_movies_ratings.csv:")
print(df_movies.columns)

print("\nColumns in final_directors_v2.csv:")
print(df_directors.columns)

# Display the first few rows of each DataFrame to understand their structure
print("\nFirst few rows of final_merged_movies_ratings.csv:")
print(df_movies.head())

print("\nFirst few rows of final_directors_v2.csv:")
print(df_directors.head())

# Merge the DataFrames on 'tconst'
final_merged_df = pd.merge(df_movies, df_directors, on='tconst', how='inner')

# Keep only the required columns: 'tconst', 'primaryTitle', 'startYear', 'genres', 'averageRating', 'director1', 'director2'
final_df = final_merged_df[['tconst', 'primaryTitle', 'startYear', 'genres', 'averageRating', 'director1', 'director2']]

# Display the final DataFrame
print("Final DataFrame:")
print(final_df.head())

# Save the final DataFrame to a new CSV file
final_df.to_csv('final_merged_movies_directors_ratings.csv', index=False)

# Download the final CSV file
files.download('final_merged_movies_directors_ratings.csv')
```

```python
from google.colab import files

# Upload the file
uploaded = files.upload()

# List the uploaded file
for filename in uploaded.keys():
    print(f'User uploaded file "{filename}" with length {len(uploaded[filename])} bytes')
```

```python
import pandas as pd

# Load the CSV file into a DataFrame
file_path = '/content/final_merged_movies_directors_ratings.csv'
df = pd.read_csv(file_path)

# Inspect the columns of the DataFrame
print("Columns in final_merged_movies_directors_ratings.csv:")
print(df.columns)

# Display the first few rows of the DataFrame to understand its structure
print("\nFirst few rows of final_merged_movies_directors_ratings.csv:")
print(df.head())

# Rename the specified columns
df.rename(columns={
    'tconst': 'imdb id',
    'primaryTitle': 'Movie title',
    'startYear': 'Year',
    'averageRating': 'IMDb score'
}, inplace=True)

# Display the final DataFrame with renamed columns
print("Final DataFrame with Renamed Columns:")
print(df.head())
```

```python
# Save the final DataFrame to a new CSV file
df.to_csv('imdb_scores.csv', index=False)

# Download the final CSV file
files.download('imdb_scores.csv')
```
